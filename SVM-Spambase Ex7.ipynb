```python
from sklearn import datasets # To Get iris dataset
from sklearn import svm # To fit the svm classifier
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt # To visuvalizing the data 
from sklearn.preprocessing import StandardScaler
group_name = 'spam'

# Loads the CSV data
df = pd.read_csv('/Users/orejoke/Downloads/spambase/spambase.data')

training_data = df.sample(frac=0.7, random_state=25)
testing_data = df.drop(training_data.index)
# The first 57 columns are features
# The last column has the correct labels (targets)
X_train, y_train = training_data.iloc[:, :57].values, training_data.iloc[:, 57].values
X_test, y_test = testing_data.iloc[:, :57].values, testing_data.iloc[:, 57].values

# Scale the features, as the original values have wide ranges
X_train = StandardScaler().fit_transform(X_train)
X_test = StandardScaler().fit_transform(X_test)
```

```python
X = X_train[:,[1,2]]
C = 1.0 # SVM regularization parameter

# SVC with linear kernel
svc = svm.SVC(kernel='linear', C=C).fit(X, y_train)
# LinearSVC (linear kernel)       
lin_svc = svm.LinearSVC(C=C).fit(X, y_train)
# SVC with RBF kernel
rbf_svc = svm.SVC(kernel='rbf', gamma=0.7, C=C).fit(X, y_train)
# SVC with polynomial (degree 3) kernel
poly_svc = svm.SVC(kernel='poly', degree=3, C=C).fit(X, y_train)
```

```python
h = .02 # step size in the mesh

# create a mesh to plot in
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
 np.arange(y_min, y_max, h))
# title for the plots
titles = ['SVC with linear kernel',
 'LinearSVC (linear kernel)',
 'SVC with RBF kernel',
 'SVC with polynomial (degree 3) kernel']


for i, clf in enumerate((svc, lin_svc, rbf_svc, poly_svc)):
 # Plot the decision boundary. For that, we will assign a color to each
 # point in the mesh [x_min, x_max]x[y_min, y_max].
 plt.subplot(2, 2, i + 1)
 plt.subplots_adjust(wspace=0.4, hspace=0.4) 

 Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])

 # Put the result into a color plot
 Z = Z.reshape(xx.shape)
 plt.contourf(xx, yy, Z, cmap=plt.cm.coolwarm, alpha=0.8)

 # Plot also the training points
 plt.scatter(X[:, 0], X[:, 1], c=y_train, cmap=plt.cm.coolwarm)
 plt.xlabel('Something1')
 plt.ylabel('Something2')
 plt.xlim(xx.min(), xx.max())
 plt.ylim(yy.min(), yy.max())
 plt.xticks(())
 plt.yticks(())
 plt.title(titles[i])

plt.show() 
```
